```python
import pandas as pd
import glob
import os
```

국민여행조사 데이터는 행이 33개인 csv와 행이 39개인 csv로 이루어짐
이에 33, 39 폴더를 만들어 따로 저장

# 33 데이터 (2023)

```python
folder_path = '/content/drive/MyDrive/2. 데이터/data/33'
all_files = glob.glob(os.path.join(folder_path, "*.csv"))
```

```python
# 2023 데이터와 2024 데이터만 가져와서 병합

dfs = []

for file in all_files:
    df = pd.read_csv(file)

    # 전체 연도 중 2023년도 추출
    df_subset = df.iloc[:, 19:36]

    # CSV 이름으로 열 이름 변경
    file_name = os.path.splitext(os.path.basename(file))[0]
    df_subset.columns = [f'{file_name}_{i+1}' for i in range(df_subset.shape[1])]

    dfs.append(df_subset)

# A, B 열은 첫 CSV에서 가져오기
index_df = pd.read_csv(all_files[0]).iloc[:, 0:2]

# 가로 합치기
combined_df = pd.concat([index_df] + dfs, axis=1)
```

```python
# 저장
output_path = '/content/drive/MyDrive/2. 데이터/data/중간 단계/combined_33_2023.csv'
combined_df.to_csv(output_path, index=False)
```

# 최종 합치기

```python
# 파일 경로
file_33_2023 = '/content/drive/MyDrive/2. 데이터/data/중간 단계/combined_33_2023.csv'
file_39_2023 = '/content/drive/MyDrive/2. 데이터/data/중간 단계/combined_39_2023_cleaned.csv'
file_33_2024 = '/content/drive/MyDrive/2. 데이터/data/중간 단계/combined_33_2024.csv'
file_39_2024 = '/content/drive/MyDrive/2. 데이터/data/중간 단계/combined_39_2024_cleaned.csv'

# CSV 불러오기
df_33_2023 = pd.read_csv(file_33_2023)
df_39_2023 = pd.read_csv(file_39_2023)
df_33_2024 = pd.read_csv(file_33_2024)
df_39_2024 = pd.read_csv(file_39_2024)

# 1, 2 열만 첫 번째 CSV에서 남기고 나머지는 제거
index_cols = df_33_2023.iloc[:, 0:2]

# 나머지 열만 선택 (가로로 합치기 위해)
df_33_2023_rest = df_33_2023.iloc[:, 2:]
df_39_2023_rest = df_39_2023.iloc[:, 2:]
df_33_2024_rest = df_33_2024.iloc[:, 2:]
df_39_2024_rest = df_39_2024.iloc[:, 2:]


combined_2023 = pd.concat([index_cols, df_33_2023_rest, df_39_2023_rest], axis=1)
combined_2024 = pd.concat([index_cols, df_33_2024_rest, df_39_2024_rest], axis=1)

# 저장
combined_2023.to_csv('/content/drive/MyDrive/2. 데이터/국민여행조사_2023.csv', index=False)
combined_2024.to_csv('/content/drive/MyDrive/2. 데이터/국민여행조사_2024.csv', index=False)
```

# 확인

```python
import pandas as pd
import re

# 확인할 파일
files = {
    "2023": '/content/drive/MyDrive/2. 데이터/국민여행조사_2023.csv',
    "2024": '/content/drive/MyDrive/2. 데이터/국민여행조사_2024.csv'
}

for year, path in files.items():
    df = pd.read_csv(path)

    cols = df.columns
    # 끝의 _숫자 제거
    clean_cols = [re.sub(r'_\d+$', '', col.strip()) for col in cols]

    # 중복 제거: 앞 부분이 같은 경우 하나만
    unique_cols = []
    seen = set()
    for col in clean_cols:
        if col not in seen:
            unique_cols.append(col)
            seen.add(col)

    # 출력
    print(f"\n{year} CSV")
    print("총 컬럼 수 (중복 제거):", len(unique_cols))
    print("컬럼 목록 (앞 부분 기준 중복 제거):")
    print(unique_cols)
    print("행 x 열:", df.shape)
```
